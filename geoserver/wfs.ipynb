```python
from owslib.wfs import WebFeatureService

# GeoServer WFS endpoint
url = "http://localhost:8888/geoserver/wfs"

# Authentication details
username = "admin"
password = "geoserver"

# Connect to the WFS service
wfs = WebFeatureService(url=url, version='1.1.0', username=username, password=password)

# Workspaces to filter
workspaces = ['sf', 'ne']

# Iterate over the available feature types and filter by workspace
for feature_type in list(wfs.contents):
    workspace, layer_name = feature_type.split(":")
    
    if workspace in workspaces:
        # Get the type of the layer
        layer_type = wfs.contents[feature_type].title
        print(f"Workspace: {workspace}, Layer Name: {layer_name}, Layer Type: {layer_type}")
```

```python
from owslib.wfs import WebFeatureService
import xml.etree.ElementTree as ET
import requests

# GeoServer WFS endpoint
url = "http://localhost:8888/geoserver/wfs"

# Authentication details
username = "admin"
password = "geoserver"

# Connect to the WFS service
wfs = WebFeatureService(url=url, version='1.1.0', username=username, password=password)

# Specify the workspace and layer you want to retrieve attributes for
workspace = 'sf'
layer_name = 'roads'
feature_type = f"{workspace}:{layer_name}"

# Retrieve the attribute table (features) for the specified layer
response = wfs.getfeature(typename=feature_type, outputFormat='application/json')

# Parse the JSON response
features = response.read()

# Optionally, you can use the `json` library to parse the JSON and handle the data
import json
feature_collection = json.loads(features)

# Print out the attributes for each feature
for feature in feature_collection['features']:
    print(f"Feature ID: {feature['id']}")
    for key, value in feature['properties'].items():
        print(f"  {key}: {value}")
    print()  # Print a newline for better readability between features
```
